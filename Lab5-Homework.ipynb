```python
print("""
    Question-1
""")

class Point():
    def __init__(self, o1, o2):
        self.op1 = o1
        self.op2 = o2
    def add(self):
        if isinstance(self.op2, Point):
            op1 = self.op1.op1 + self.op2.op1
            op2 = self.op1.op2 + self.op2.op2
            n_p = Point(op1,op2)
            return n_p
        elif isinstance(self.op2, tuple):
            x = self.op1.op1 + self.op2[0]
            y = self.op1.op2 + self.op2[1]
            n_p_t = Point(x,y)
            return n_p_t

x = Point(1,2)
x1 = Point(3,4)
p = Point(x, x1)
r = p.add()
print(r.op1, r.op2)
t = (0,2)
p1 = Point(x, t)
r1 = p1.add()
print(r1.op1, r1.op2)
```
